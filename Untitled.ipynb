```python
def download(url):
    split = urlparse.urlsplit(url)
    filename = "Images/" + split.path.split("/")[-1]
    urllib.urlretrieve(url, filename)
    img = graphlab.Image(filename)
    return img
```

```python
array  = Images['ITEMIMAGE_IMAGE'].apply(download)
```

```python
def intable(url):
    split = urlparse.urlsplit(url)
    filename = "Images/" + split.path.split("/")[-1]
    img = graphlab.Image(filename)
    return img
```
